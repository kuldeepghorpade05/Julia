```julia
Pkg.add("Optim")
```

```julia
import Optim
f(x) = (1.0 - x[1])^2 + 100.0 * (x[2] - x[1]^2)^2
```

```julia
optimize(f, [0.0, 0.0])
```

```julia
optimize(f, [0.0, 0.0], NelderMead())
```

```julia
optimize(f, [0.0, 0.0], LBFGS())
```
