```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

```julia
plotly()
```

```julia
train=CSV.read("loan.csv",DataFrame,normalizenames=true)
```

```julia
train=dropmissing(train)
```

```julia
Plots.histogram(train[!,"ApplicantIncome"],bins=50,xlabel="ApplicantIncome",labels="Frequen")
```

```julia
using StatsPlots
StatsPlots.boxplot(train[!,"ApplicantIncome"],xlabel="ApplicantIncome")
```

```julia
using StatsPlots
StatsPlots.boxplot(train[!,"ApplicantIncome"],
 xlabel="ApplicantIncome")
```

```julia
using StatsPlots
StatsPlots.scatter(train[!,"LoanAmount"], train[!,"ApplicantIncome"],
 xlabel="LoanAmount",ylabel="ApplicantIncome")
```
