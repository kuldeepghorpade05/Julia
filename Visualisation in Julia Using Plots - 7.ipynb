```julia

Plots.histogram(train[!,"ApplicantIncome"],bins=50,xlabel="ApplicantIncome",labels="frequency")
```

```julia
using StatsPlots
StatsPlots.boxplot(train[!,"ApplicantIncome"], xlabel="ApplicantIncome")
```

```julia
StatsPlots.boxplot(train[!,"Education"],train[!,"ApplicantIncome"],labels="Applicantincome")
```

```julia
Plots.histogram(train[!,"LoanAmount"],bins=50,xlabel="LoanAmount",labels="Frequency")
```

```julia
using StatsPlots
StatsPlots.boxplot(train[!,"LoanAmount"], ylabel="LoanAmount")
```
